```python
"""Déjà nous voyons que la colonne year n'est pas dans le format d'année (Que nous allons convertir), il y a des données 
manquantes également à corriger dans le tableau. Pour simplifier l'analyse nous allons supprimer toutes les lignes qui 
contiennent des cellules vides et extraire les données concernant le pays qui dispose de plus de données pour l'analyse.
"""

df=df.dropna()
```

```python
df["country"].value_counts()
```

```python
"""Nous voyons que la France dispose de plus de données alors nous alons étudier sa situation. Nous n'avons pas besoin de la 
tésorerie des USA dans cas donc nous allons supprimer cette colonne. 
"""

df_France = df.groupby("country").get_group("France").drop("USTreasury", axis=1)

df_France.head()
```

```python
x=df_France["year"]
y=df_France["manufacturingoutput"]
```

```python
"""Nous allons faire une représentation graphique pour voir l'évolution du taux d'inflation, du taux de change, de la production 
nationale, le prix du carburant, au fils des années de la France au vue des données du tableau"""


plt.style.use("ggplot")

plt.figure(figsize=(12,5))
plt.plot(x,y, color="blue", label= "PRODUCTION NATIONALE" )


plt.title("EVOLUTION DE LA PRODUCTION NATIONALE")
plt.xlabel("ANNEE")
plt.ylabel("PRODUCTION NATIONALE")
plt.legend()
plt.show()
```
